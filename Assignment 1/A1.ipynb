```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

# 1. Load a color and grayscale image

```python
color_img = cv2.imread('A1.jpeg', cv2.IMREAD_COLOR)
gray_img = cv2.imread('A1.jpeg', cv2.IMREAD_GRAYSCALE)
```

# 2. Display using OpenCV

```python
cv2.namedWindow('Coloured LH', cv2.WINDOW_NORMAL)      # Create resizable window
cv2.resizeWindow('Coloured LH', 500, 400)              # Set size
cv2.imshow('Coloured LH', color_img)                   # Show color image

cv2.namedWindow('Gray LH', cv2.WINDOW_NORMAL)
cv2.resizeWindow('Gray LH', 500, 400)
cv2.imshow('Gray LH', gray_img)                        # Show grayscale image

cv2.waitKey(0)                                         # Wait for any key
cv2.destroyAllWindows()                                # Close all windows
```

# Display using Matplotlib (convert BGR to RGB for color)

```python
color_img_rgb = cv2.cvtColor(color_img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(color_img_rgb)
plt.title('Color Image - Matplotlib')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gray_img, cmap='gray')
plt.title('Grayscale Image - Matplotlib')
plt.axis('off')

plt.tight_layout()
plt.show()
```

# 3. Print image properties

```python
print("Color Image Shape:", color_img.shape)
print("Color Image Dtype:", color_img.dtype)
print("Color Image Pixel Range:", color_img.min(), "-", color_img.max())

print("Grayscale Image Shape:", gray_img.shape)
print("Grayscale Image Dtype:", gray_img.dtype)
print("Grayscale Image Pixel Range:", gray_img.min(), "-", gray_img.max())
```

# 4. Change pixel at (50, 50) to red and 50x50 block to red

```python
color_img[50, 50] = [0, 0, 255]  # BGR red pixel
color_img[0:50, 0:50] = [0, 0, 255]  # Top-left block
```

```python
cv2.namedWindow('Edit LH', cv2.WINDOW_NORMAL)
cv2.resizeWindow('Edit LH', 500, 400)
cv2.imshow('Edit LH', color_img)                        

cv2.waitKey(0)                                         
cv2.destroyAllWindows()        
```

# 5. Manual grayscale conversion


```python
manual_gray = np.mean(color_img, axis=2).astype(np.uint8)
```

```python
cv2.namedWindow('Manual LH', cv2.WINDOW_NORMAL)
cv2.resizeWindow('Manual LH', 500, 400)
cv2.imshow('Manual LH', manual_gray)                        

cv2.waitKey(0)                                         
cv2.destroyAllWindows() 
```

# 6. Split channels


```python
B, G, R = cv2.split(color_img)

cv2.imshow('Blue Channel', B)
cv2.imshow('Green Channel', G)
cv2.imshow('Red Channel', R)

cv2.waitKey(0)                                         
cv2.destroyAllWindows() 
```

# 7. Save modified image


```python
cv2.imwrite('modified_image.jpg', color_img)
```

# 8. Observation - compare grayscale methods


```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(gray_img, cmap='gray')
plt.title('OpenCV Grayscale')

plt.subplot(1, 2, 2)
plt.imshow(manual_gray, cmap='gray')
plt.title('Manual Grayscale (mean)')

plt.tight_layout()
plt.show()

# Wait for key press and close OpenCV windows
cv2.waitKey(0)
cv2.destroyAllWindows()
```
